```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Distribution of 2022 Population
plt.figure(figsize=(10, 6))
sns.histplot(df['2022 Population'], bins=50, kde=True)
plt.title('Distribution of 2022 Country Populations')
plt.xlabel('2022 Population')
plt.ylabel('Number of Countries')
plt.xscale('log') # Use log scale due to high skewness
plt.show()

# Example: Distribution of Growth Rate
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Growth Rate'])
plt.title('Distribution of Country Growth Rates')
plt.xlabel('Growth Rate')
plt.show()
```

```python
# Example: Total Population by Continent
plt.figure(figsize=(12, 7))
continent_population = df.groupby('Continent')['2022 Population'].sum().sort_values(ascending=False)
sns.barplot(x=continent_population.index, y=continent_population.values)
plt.title('Total Population by Continent (2022)')
plt.xlabel('Continent')
plt.ylabel('Total Population')
plt.ticklabel_format(style='plain', axis='y') # Prevent scientific notation on y-axis
plt.show()

```

```python
# Example: Growth Rate Distribution by Continent
plt.figure(figsize=(12, 7))
sns.boxplot(x='Continent', y='Growth Rate', data=df)
plt.title('Population Growth Rate Distribution by Continent')
plt.xlabel('Continent')
plt.ylabel('Growth Rate')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Example: 2022 Population vs. Area (Km²)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Area (km²)', y='2022 Population', data=df, hue='Continent', size='Density (per km²)', sizes=(20, 400))
plt.title('2022 Population vs. Area (Km²) by Continent')
plt.xlabel('Area (Km²)')
plt.ylabel('2022 Population')
plt.xscale('log') # Log scale for better spread if values vary widely
plt.yscale('log') # Log scale for better spread if values vary widely
plt.show()
```

```python
# To plot global population trend, first aggregate data
population_columns = [col for col in df.columns if 'Population' in col and 'World Population Percentage' not in col]
df_population_long = df[population_columns].sum().reset_index()
df_population_long.columns = ['Year', 'Total Population']
df_population_long['Year'] = df_population_long['Year'].str.extract('(\d{4})').astype(int)
df_population_long = df_population_long.sort_values('Year')

plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Total Population', data=df_population_long)
plt.title('Global Population Trend (1970-2022)')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.grid(True)
plt.ticklabel_format(style='plain', axis='y')
plt.show()

```

#Global Population is rising linearly since 1970


```python

# Select only numerical columns for correlation calculation
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
plt.figure(figsize=(14, 12))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
